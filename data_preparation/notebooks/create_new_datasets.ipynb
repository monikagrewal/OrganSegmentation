# Introduction
This notebook was used to
* create a train/test split of the raw data based on the old train/test, and the presence of cervical cancer
* copy all the raw dicom files to new train/test folders

```python
import pandas as pd
```

## Create new dataset split

```python
# dd = pd.read_excel('/export/scratch3/grewal/Data/data_details.xlsx')
ddf = pd.read_csv('/export/scratch3/grewal/Data/data_details_final.csv')
ddf = ddf.assign(
    ThreeD_old=ddf.ThreeD,
    ThreeD=((ddf.ThreeD == 'test') & (ddf.CervicalCancer == 1)).replace({True: 'test', False: 'train'})
)
ddf.sample(10)
```

```python
ddf.to_csv('/export/scratch2/bvdp/Data/data_details_25-06-2020.csv', index=False)
```

## Copy data based on new split

```python
import os
import shutil
```

```python
df = pd.read_csv('/export/scratch2/bvdp/Data/data_details_25-06-2020.csv')
df.head(3)
```

### Copy dicom files

```python
old_dir = '/export/scratch2/grewal/Data/Projects_DICOM_data/ThreeD/MODIR_data_{}_split'
new_dir = '/export/scratch2/bvdp/Data/Projects_DICOM_data/ThreeD/MODIR_data_{}_split'
```

```python
for idx, row in df.iterrows():
    old_path = os.path.join(old_dir.format(row.ThreeD_old), row.PatientID)
    new_path = os.path.join(new_dir.format(row.ThreeD), row.PatientID)
    print(old_path)
    print(new_path, end='\n\n')    
#     os.mkdir(dest)
#     dest_dir = os.path.join(dest,os.path.basename(source))
    shutil.copytree(old_path, new_path)
```

### Copy jpg's

```python
old_dir_jpg = '/export/scratch3/grewal/Data/Projects_JPG_data/ThreeD/MODIR_data_{}_split'
new_dir_jpg = '/export/scratch2/bvdp/Data/Projects_JPG_data/ThreeD/MODIR_data_{}_split'
```

```python
for idx, row in df.query('ThreeD == "test"').iterrows():
    print(idx, end=',')
    old_path = os.path.join(old_dir_jpg.format(row.ThreeD_old), row.PatientID)
    new_path = os.path.join(new_dir_jpg.format(row.ThreeD), row.PatientID)
#     print(old_path)
#     print(new_path, end='\n\n')    
    if not os.path.exists(old_path):
        print('skipping')
        continue    

    shutil.copytree(old_path, new_path)
```
